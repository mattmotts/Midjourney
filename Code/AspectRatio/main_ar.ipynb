```python
# creates a 4:3 and 5:4 aspect ratio image from the 3:2 upscaled 

import os
from PIL import Image

def resize_image(image_path, aspect_ratio, new_name):
    with Image.open(image_path) as img:
        # Get the original dimensions
        width, height = img.size
        
        # Calculate new dimensions
        if aspect_ratio == (4, 3):
            new_width = width
            new_height = int(width * 3 / 4)
        elif aspect_ratio == (5, 4):
            new_width = width
            new_height = int(width * 4 / 5)
        else:
            raise ValueError("Unsupported aspect ratio")
        
        # Resize the image
        resized_img = img.resize((new_width, new_height), Image.LANCZOS)
        
        # Save the resized image
        resized_img.save(new_name)
        print(f"Saved resized image: {new_name}")

def process_images(directory):
    for root, dirs, files in os.walk(directory):
        for file in files:
            # Check if the file is a 12x PNG image
            if file.endswith("12x.png"):
                # Construct the full file path
                file_path = os.path.join(root, file)
                
                # Get the base name without extension
                base_name = os.path.splitext(file)[0]
                
                # Create new names for the resized images
                new_name_43 = os.path.join(root, f"{base_name}_43.png")
                new_name_54 = os.path.join(root, f"{base_name}_54.png")
                
                # Resize and save the images
                resize_image(file_path, (4, 3), new_name_43)
                resize_image(file_path, (5, 4), new_name_54)
                
                # Rename the original 12x image to include an underscore before 12x
                new_name_12x = os.path.join(root, f"{base_name}_12x.png")
                os.rename(file_path, new_name_12x)
                print(f"Renamed original image: {new_name_12x}")

# Specify the directory you want to process
directory_path = '/Users/matt/Pictures/Midjourney/Listings'

# Call the function to process images
process_images(directory_path)
```
